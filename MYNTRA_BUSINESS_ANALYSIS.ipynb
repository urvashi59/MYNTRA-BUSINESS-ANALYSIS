# ‚è© FIX "DiscountOffer" COLUMNS

```python
import pandas as pd

# Assuming your DataFrame is named 'myntra_copy'
myntra_copy['DiscountOffer'] = myntra_copy['DiscountOffer'].str.extract('(\d+)')
myntra_copy['DiscountOffer'] = pd.to_numeric(myntra_copy['DiscountOffer'], errors='coerce') # Convert to numeric, handling errors

myntra_copy = myntra_copy[myntra_copy['DiscountOffer'] < 100]
myntra_copy.dropna(subset=['DiscountOffer'], inplace=True)
myntra_copy['DiscountOffer'] = myntra_copy['DiscountOffer'].astype(int)
```

```python
myntra_copy.info()
```

```python
myntra_copy["DiscountOffer"].value_counts()
```

```python
myntra_copy.isnull().sum()
```

# ‚è© FILL MISSING VALUE IN "DiscountPrice (in Rs)" COLUMNS

```python
missing_discount_price_mask = myntra_copy['DiscountPrice (in Rs)'].isnull()
has_discount_offer_mask = myntra_copy['DiscountOffer'].notnull()

rows_to_fill = missing_discount_price_mask & has_discount_offer_mask

myntra_copy.loc[rows_to_fill, 'DiscountPrice (in Rs)'] = (
    myntra_copy.loc[rows_to_fill, 'OriginalPrice (in Rs)'] *
    (1 - myntra_copy.loc[rows_to_fill, 'DiscountOffer'] / 100)
).round(2)  # Round to 2 decimal places
```

```python
print("Remaining nulls in DiscountPrice (in Rs):",
         myntra_copy['DiscountPrice (in Rs)'].isnull().sum())
```

```python
myntra_copy.isnull().sum()
```

```python
myntra_copy["DiscountPrice (in Rs)"].value_counts()
```

# **üß† STEP 3: FEATURE ENGINEERING & ENHANCEMENT**



*   This step is all about creating new, meaningful features and refining existing ones to boost model performance and data understanding.



# ‚è© TOTAL NUMBER OF SIZE OPTIONS

```python
# TOTAL NUMBER OF SIZE OPTIONS

myntra_copy['NumSizes'] = myntra_copy['SizeOption'].apply(lambda x: len(str(x).split(',')))
myntra_copy.head()
```

# ‚è© BRAND POPULARITY (TOP BRANDS)

```python
#BRAND POPULARITY (TOP BRANDS)

top_brands = myntra_copy['BrandName'].value_counts().nlargest(10).index
myntra_copy['TopBrandFlag'] = myntra_copy['BrandName'].apply(lambda x: 1 if x in top_brands else 0)
myntra_copy.head()
```

# ‚è© CATEGORY-BASED DISCOUNT MEAN

```python
category_discount_avg = myntra_copy.groupby('Category')['DiscountOffer'].transform('mean')
myntra_copy['CategoryDiscountAvg'] = category_discount_avg.round(2)
myntra_copy.head()
```

# ‚è© DISCOUNT INTENSITY CATEGORY

```python
def discount_level(x):
    if x < 30:
        return 'Low'
    elif 30 <= x < 60:
        return 'Medium'
    else:
        return 'High'

myntra_copy['DiscountLevel'] = myntra_copy['DiscountOffer'].apply(discount_level)
myntra_copy.head()
```

# ‚è© HASRATING AND HASREVIEW

```python
# Create binary features as 'Yes' or 'No'
myntra_copy['HasRating'] = myntra_copy['Ratings'].notnull().map({True: 'Yes', False: 'No'})
myntra_copy['HasReview'] = myntra_copy['Reviews'].notnull().map({True: 'Yes', False: 'No'})

# Drop the original columns
myntra_copy.drop(['Ratings', 'Reviews'], axis=1, inplace=True)

# Quick check
myntra_copy[['HasRating', 'HasReview']].head()
```

```python
myntra_copy.sample(5)
```

# ‚è© TOP INDIVIDUAL CATEGORY

```python
top_individual_categories = myntra_copy['Individual_category'].value_counts().nlargest(10).index
myntra_copy['TopIndCategoryFlag'] = myntra_copy['Individual_category'].apply(lambda x: 1 if x in top_individual_categories else 0)

myntra_copy.head()
```
